## Graphs

```python
import matplotlib.pyplot as plt
```

### Bar

```python
#Y values
y_values = (2, 5)
#x values and axis
x_values = (0, 1)
x_labels = ("Pass", "Fail")
plt.xticks(x_values, x_labels)
#Plot the bars
plt.bar(x_values, y_values, width=0.9, color=('g','r'))
#Labels
plt.title("Results")
plt.ylabel("Count")
plt.xlabel("Result")
#Display
plt.show()

```

### Pie
Using same  values as bar

```python
plt.pie(y_values, labels=x_labels, autopct='%1.1f%%',colors=('g','r'))
plt.axis('equal')
plt.show()
```

### Combined figure

Two graphs as subplots of the same figure

```python
#Y values
y_values = (2, 5, 3)
#x values and axis
x_values = (0, 1, 2)
x_labels = ("Pass", "Blocked", "Fail")
cols = ('r','b','g')

#Create figure (figsize is optional)
plt.figure(1, figsize=(10, 5))

#Bar Chart
plt.subplot(121)#number of rows, number of columns, position
plt.xticks(x_values, x_labels)
plt.bar(x_values, y_values, width=0.9, color=cols)
plt.ylabel("Count")
plt.xlabel("Result")

#Pie Chart
plt.subplot(122)
plt.pie(y_values, labels=x_labels, autopct='%1.1f%%',colors=cols)
plt.axis('equal')

#Main title
plt.suptitle('Main title')

#Reduce overlap
##plt.tight_layout(h_pad=2.5)

#Improve spacing
##plt.subplots_adjust(wspace = 0.4) 

plt.show()
```
